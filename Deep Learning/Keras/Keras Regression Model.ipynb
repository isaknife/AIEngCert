# Part A: Build baseline model

Imports

```python
import pandas as pd
import numpy as np
import keras
from keras.models import Sequential
from keras.layers import Dense
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
import statistics
```

```python
concrete_data = pd.read_csv('https://s3-api.us-geo.objectstorage.softlayer.net/cf-courses-data/CognitiveClass/DL0101EN/labs/data/concrete_data.csv')
concrete_data.head()
```

Target: Strength

Predictors: Cement, Blast Furnace Slag, Fly Ash, Water, Superplasticizer, Coarse Aggregate, Fine Aggregate

Split data into predictors and target

```python
concrete_data_columns = concrete_data.columns

X = concrete_data[concrete_data_columns[concrete_data_columns != 'Strength']] # all columns except Strength
Y = concrete_data['Strength'] # Strength column
```

Sanity Check

```python
X.head()
```

```python
Y.head()
```

```python
n_cols = X.shape[1] # number of predictors
```

Define regression model w/ one hidden layer of 10 nodes, and a ReLU activation function

```python
def regression_model():
    # create model
    model = Sequential()
    model.add(Dense(10, activation='relu', input_shape=(n_cols,)))
    model.add(Dense(1))
    
    # compile model
    model.compile(optimizer='adam', loss='mean_squared_error')
    return model
```

Split training and test, using 30% for testing

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.3)
```

Train the model using 50 epochs

```python
model = regression_model()
model.fit(X_train, Y_train, validation_split=0.3, epochs=50)
```

Get predictions on X_test

```python
Y_predicted = model.predict(X_test)
```

Calculate MSE

```python
mse = mean_squared_error(Y_test, Y_predicted)
mse
```

Repeat 50 times, report mean and sd of MSE

```python
mses = []
for i in range(50):
    #split
    X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.3)
    #train
    model = regression_model()
    model.fit(X_train, Y_train, validation_split=0.3, epochs=50)
    #predict
    Y_predicted = model.predict(X_test)
    #calculate MSE
    mse = mean_squared_error(Y_test, Y_predicted)
    #add to list
    mses.append(mse)
```

```python
mses
```

Mean

```python
statistics.mean(mses)
```

Standard deviation

```python
statistics.pstdev(mses)
```

# Part B: Normalize data

```python
X_norm = (X - X.mean()) / X.std()
X_norm.head()
```

```python
mses_b = []
for i in range(50):
    #split
    X_train, X_test, Y_train, Y_test = train_test_split(X_norm, Y, test_size=0.3)
    #train
    model = regression_model()
    model.fit(X_train, Y_train, validation_split=0.3, epochs=50)
    #predict
    Y_predicted = model.predict(X_test)
    #calculate MSE
    mse = mean_squared_error(Y_test, Y_predicted)
    #add to list
    mses_b.append(mse)
```

```python
mses_b
```

```python
statistics.mean(mses_b)
```

```python
statistics.pstdev(mses_b)
```

# Part C: Increase epochs

Increase epochs from 50 to 100

```python
mses_c = []
for i in range(50):
    #split
    X_train, X_test, Y_train, Y_test = train_test_split(X_norm, Y, test_size=0.3)
    #train
    model = regression_model()
    model.fit(X_train, Y_train, validation_split=0.3, epochs=100)
    #predict
    Y_predicted = model.predict(X_test)
    #calculate MSE
    mse = mean_squared_error(Y_test, Y_predicted)
    #add to list
    mses_c.append(mse)
```

```python
statistics.mean(mses_c)
```

```python
statistics.pstdev(mses_c)
```

# Part D: Increase hidden layers

Use 3 hidden layers each of 10 nodes and a ReLU activation function

```python
def regression_model_d():
    # create model
    model_d = Sequential()
    model_d.add(Dense(10, activation='relu', input_shape=(n_cols,)))
    model_d.add(Dense(10, activation='relu'))
    model_d.add(Dense(10, activation='relu'))
    model_d.add(Dense(1))
    
    # compile model
    model_d.compile(optimizer='adam', loss='mean_squared_error')
    return model_d
```

```python
mses_d = []
for i in range(50):
    #split
    X_train, X_test, Y_train, Y_test = train_test_split(X_norm, Y, test_size=0.3)
    #train
    model_d = regression_model_d()
    model_d.fit(X_train, Y_train, validation_split=0.3, epochs=100)
    #predict
    Y_predicted = model_d.predict(X_test)
    #calculate MSE
    mse = mean_squared_error(Y_test, Y_predicted)
    #add to list
    mses_d.append(mse)
```

```python
mses_d
```

```python
statistics.mean(mses_d)
```

```python
statistics.pstdev(mses_d)
```

# Final Results

```python
print('Part A:\nMean:',statistics.mean(mses),'\nStdev:', statistics.pstdev(mses))
print('\nPart B:\nMean:',statistics.mean(mses_b),'\nStdev:', statistics.pstdev(mses_b))
print('\nPart C:\nMean:',statistics.mean(mses_c),'\nStdev:', statistics.pstdev(mses_c))
print('\nPart D:\nMean:',statistics.mean(mses_d),'\nStdev:', statistics.pstdev(mses_d))
```
